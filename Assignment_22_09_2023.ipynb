```python
mi_q1 = df["MonthlyIncome"].quantile(0.25)
mi_q3 = df["MonthlyIncome"].quantile(0.75)
mi_iqr = mi_q3 - mi_q1

mi_ul = mi_q3 + 1.5*mi_iqr
mi_ul
```

```python
df["MonthlyIncome"][df["MonthlyIncome"]>mi_ul].count()
```

```python
mi_median = df["MonthlyIncome"].median()
mi_median
```

```python
df["MonthlyIncome"] = np.where(df["MonthlyIncome"]>mi_ul,
                               mi_median, df["MonthlyIncome"])
```

```python
sns.boxplot(df["MonthlyIncome"])
```

# **Feature Scaling**

---



```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
```

```python
X_scaled = pd.DataFrame(ms.fit_transform(X), columns = X.columns)
X_scaled.head()
```

# **Splitting into Training & Testing Set**

---



```python
from sklearn.model_selection import train_test_split
```

## **Splitting Unscaled Set**

---


For Decision Tree & Random Forest Classification

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y,
                                      test_size = 0.2, random_state = 0)
```

```python
print(X_train1.shape)
print(X_test1.shape)
print(y_train1.shape)
print(y_test1.shape)
```

## **Splitting Scaled Set**

---



```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_scaled, y,
                                      test_size = 0.2, random_state = 0)
```

```python
print(X_train2.shape)
print(X_test2.shape)
print(y_train2.shape)
print(y_test2.shape)
```

# **Model Building**

---



## **Logistic Regression**

---



```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

### **Training the Model**

---


```python
lr_model.fit(X_train2,)
```

### **Making Predictions**

---



```python
y_preds1 = lr_model.predict(X_test2)
```

### **Evaluation of the Model**

---



```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
acc1 = accuracy_score(y_test2, y_preds1)
acc1
```

```python
confusion_matrix(y_test2, y_preds1)
```

```python
pd.crosstab(y_test2, y_preds1)
```

```python
print(classification_report(y_test2, y_preds1))
```

```python
roc_auc_score(y_test2, y_preds1)
```

```python
prob1 = lr_model.predict_proba(X_test2)[:,1]
```

```python
fpr1, tpr1, thres1 = roc_curve( y_test2, prob1)
```

```python
plt.plot(fpr1, tpr1)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.show()
```

### **Hyperparameter Tuning**

---



```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
 'criterion':['gini','entropy'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5,7,8,9],
  'max_features':['auto', 'sqrt', 'log2']
}
```

```python
grid_search = GridSearchCV(estimator=dtc_model, param_grid = params,
             cv=5, scoring = "accuracy")
```

```python
grid_search.fit(X_train1, y_train1)
```

```python
grid_search.best_params_
```

```python
dtc_cv = DecisionTreeClassifier(criterion= 'entropy', max_depth= 3,
                                max_features= 'sqrt', splitter= 'best')
```

```python
dtc_cv.fit(X_train1, y_train1)
```

```python
y_preds2 = dtc_cv.predict(X_test1)
```

```python
print(classification_report(y_test1, y_preds2))
```

```python
acc2 = accuracy_score(y_test1, y_preds2)
acc2
```

## **Random Forest Classifier**

```python
rfc_cv = GridSearchCV(estimator=rfc_model, param_grid = rfc_params,
                      cv=10, scoring = "accuracy")
```

```python
rfc_cv.best_params_
```
